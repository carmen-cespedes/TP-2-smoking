# 01 - Lectura y Discovery
Carga inicial del dataset etiquetado y exploración de su estructura.

```python
import ssl
import certifi
import pandas as pd

# Crear contexto SSL con certificados de certifi
ssl_context = ssl.create_default_context(cafile=certifi.where())

import urllib.request
url = f'https://docs.google.com/spreadsheets/d/{SHEET_ID_LABELED}/export?format=csv'
with urllib.request.urlopen(url, context=ssl_context) as response:
    df = pd.read_csv(response)

print(f"Shape: {df.shape}")
df.head()
```

```python

# Información general del dataset
df.info()
```

```python

# Estadísticas descriptivas
df.describe()
```

```python

# Valores nulos por columna
null_summary = df.isnull().sum().reset_index()
null_summary.columns = ['columna', 'nulos']
null_summary['pct'] = (null_summary['nulos'] / len(df) * 100).round(2)
null_summary[null_summary['nulos'] > 0].sort_values('pct', ascending=False)
```

```python

# Distribución del target
print("Distribución de la variable objetivo 'smoking':")
print(df['smoking'].value_counts())
print(f"\nBalance: {df['smoking'].mean():.2%} fumadores")
```

```python

# Tipos de datos
print("Columnas numéricas:", df.select_dtypes(include='number').columns.tolist())
print("\nColumnas categóricas:", df.select_dtypes(include='object').columns.tolist())
```

```python

# Guardar copia raw
df.to_csv('../data/raw/smoking_labeled.csv', index=False)
print("Dataset raw guardado en data/raw/smoking_labeled.csv")
```
